```python
import matplotlib.pyplot as plt
import numpy as np

custom_preamble = {
    "text.usetex": True,
    "text.latex.preamble": r"\usepackage{amsmath} \usepackage{amsfonts}"
    }
plt.rcParams.update(custom_preamble)
```

```python
def plot_riemann(f = lambda x: x**2, amnt = 10, title = '$f(x) = x^2$'):
    fig, ax = plt.subplots(figsize=(7,5))

    if amnt == 0:
        x = np.arange(0, 1, 0.01)
        y = f(x)
        ax.plot(x, y, color="black")
        ax.set_title(title, fontsize='xx-large')
        return


    x = np.arange(0, 1, 0.01)
    part = np.arange(0, 1, 1 / amnt)
    y = f(x)
    part_y = f(part)

    upper_sum = sum(1 / amnt * f(part))
    lower_sum = sum(1 / amnt * np.append(part_y[1:],f(1)))

    # ax.set_xticks([])
    # ax.set_yticks([])
    ax.plot(x, y, color="black")
    ax.bar(part, np.append(part_y[1:],f(1)), align="edge", width=1/amnt, linewidth=1-amnt/200, edgecolor="black", color='lightcoral')
    ax.bar(part, part_y, align="edge", width=1/amnt, linewidth=1-amnt/200, edgecolor="black", color='cornflowerblue')
    # ax.text(0, 1.1, '$f(x) = x^2$', fontsize='xx-large')
    ax.set_title(title, fontsize='xx-large')
    ax.text(0.1,0.7,(rf'\begin{{align*}}'
                     rf'&U(f, P) = {upper_sum:.4}\\'
                     rf'&L(f, P) = {lower_sum:.4}'
                     rf'\end{{align*}}'),
                    fontsize="xx-large")

plot_riemann(lambda x: x**2, 0)
```
